### **Markov Decision Processes**

* Model RL environment mathematically
    - Simlify the complex environement
* We define the complex environment in the form of 
    - States
    - Actions
    - Rewards
    - &rarr;Transition probablities
    - &rarr; Which we solve with model based RL techniques.

### **Markov Property**
* Future state depends only on current state and action
   - Just like in chess game

*eg*:\
**Frozen Lake as MDP-states**
* Postion agent can occupy\
**Frozen Lake as MDP-terminations_states**
* Lead to episode termination\
**Frozen Lake as MDP-actions**
* Up,down,right,left\
**Frozen Lake as MDP-transitions**
* Actions doesn't necessarily leads to expected outcomes\
**Frozen Lake as MDP-rewards**
* Reward only given in goal state

````python
import gymnasium as gym
````

````python
env=gym.make('FrozenLake',is_slippery=True)#sliipery cause the unexpected movements causing unexpected perpendicular movement instead of the left
print(env.action_space)#prints the actions that can happed
print(env.observation_space)#prints the number of the possible postions that a agent can take, in this case 16
````

### **Gymnasium rewards and transitions**
**```env.unwrapped.P```**: dictionary where keys are state-action pairs.\
```print(env.unwrapped.P[state][action])```\
[\
    (probablity_1,next_state_1,reward_1,is_terminal_1),\
    (probablity_2,next_state_2,reward_2,is_terminal_2)\
]

````python
state=6#indicating the box postion counting starting from zero
action=0#moving left
print(env.unwrapped.P[state][action])
````

**Interpretation of the resuts above**\
* These are the tuples each representing the potential outcomes. Here for the first tuple, there is the probablity of moving down which is 33.33%. And there  is no terminating case there. And the reward for this is zero. The second element in the tuple represents the next move to box postion 2.

````python
# Create the Cliff Walking environment
env = gym.make('CliffWalking')

# Compute the size of the action space
num_actions = env.action_space.n

# Compute the size of the state space
num_states = env.observation_space.n

print("Number of actions:", num_actions)
print("Number of states:", num_states)
````

````python
env=gym.make('CliffWalking',render_mode='rgb_array')
state, info = env.reset(seed=42)
state_image = env.render()
import matplotlib.pyplot as plt
plt.imshow(state_image)
plt.show()
````

````python
# Choose the state
state = 6

# Extract transitions for each state-action pair
for action in range(num_actions):
    transitions = env.unwrapped.P[state][action]
    # Print details of each transition
    for transition in transitions:
        probability,next_state,reward, done = transition
        print(f"Probability: {probability}, Next State: {next_state}, Reward: {reward}, Done: {done}")
````
